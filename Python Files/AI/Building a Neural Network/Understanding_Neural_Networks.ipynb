There are six attributes of a neural network:

1. Input layer of neurons
2. Some number of hidden layers of neurons
3. Output layer of neurons connecting to the world
4. Set of weights and biases between each neuron level
5. Choice of activation function for each hidden layer of neurons
6. Choice of loss function to reduce overtraining the network

A neuron, in AI terms, is a software model of a nervous system cell that behaves 
more or less like a brain neuron. The model uses numbers to make one neuron or
another more important to the results. These are called "weights"

#Layers of neurons

The neural network recieves an example, and guesses at the naswer (by using whatever default weights and biases that it starts off with.) if the answer is wrong, it backtracks and then modifies the weights and biases in the neurons, in an effort to fix the error. This is called **backpropagation** and simulates what people do when performing any given task, trial-and-error.

After you do the same process many times, the neural network begins to get better and learn, it provides better answers with each iteration. The process can take days, sometimes weeks and a lot of computer power to complete tasks.

Even though it may take an extensive amount of time to train a neural net, after it's trained it can easily be duplicated by copying the topology, weights and the biases of the trained network. This means that you can use it again and again (while not using so much computer power).

Neural networks do model some types of human learning, but humans have significantly more complex ways to categorize objects hierarchically with little effort (horses and pigs as animals for example). Neural networks & the deep learning field are not quite as good at transferring knowledge and results without restraining.
